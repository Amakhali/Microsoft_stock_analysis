```python
# Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download Microsoft stock data
data = yf.download("MSFT", start="2010-01-01", end="2025-01-01")

# Save data to CSV
#data.to_csv("../data/msft_stock_data.csv")

# Plot closing prices
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='MSFT Closing Price')
plt.title('Microsoft Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```
